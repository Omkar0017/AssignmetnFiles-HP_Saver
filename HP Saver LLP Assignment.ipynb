## Code Explanation


1. First we read example Technical Skills file which will be used for training our model
2. We create list named as exTechlst which consists of list of all the technical skills used for training.
3. Since we only have positive values i.e only technical skills we read another text file to get non-technical values so we can properly train our model.
4. We need to process the values since they are continious sentences.  At first we apply work tokenizer to get seperate words and then we apply port Stemmer to extract root words.
5. we form another array of 1's and 0's so we can have target variable.
6. In X we have technical and non-Technical skills and in Y we have 1's and 0's
7. We split them in training and testing set of 80% and 20% respectively
8. We use GaussianNB classifier to classify the data.
9. Since we can't feed words directly to the model we use CountVectorizer to convert words into vectors.
10. After this process we train the model and check accuracy
11. After that we apply final data i.e. Raw_Skills_Dataset to model and classify it into 2  parts.

```python
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import  word_tokenize
```

```python
import csv
exTech = []
with open("Example_Technical_Skills.csv", "r") as csvfile:
    reader_variable = csv.reader(csvfile, delimiter=",")
    for row in reader_variable:
        #print(row)
        exTech.append(row)
        
rawSkills = []
with open("Raw_Skills_Dataset.csv", "r",encoding='utf-8') as csvfile:
    reader_variable = csv.reader(csvfile, delimiter=",")
    for row in reader_variable:
        #print(row)
        rawSkills.append(row)
```

```python
exTechlst = []
rawSkillslst = []

for i in exTech:
    for j in i:
       exTechlst.append(j) 
    
for i in rawSkills:
    for j in i:
       rawSkillslst.append(j)
```

```python
exTechlst
```

```python
rawSkillslst
```

```python
stop_words = list(stopwords.words('english'))
stop_words.sort()
stop_words
```

```python
NonTech = []
with open("non_technicalSkills.txt", "r") as csvfile:
    reader_variable = csv.reader(csvfile, delimiter=",")
    for row in reader_variable:
        #print(row)
        NonTech.append(row)
        
NonTechlst = []
for i in NonTech:
    for j in i:
       NonTechlst.append(j) 
    
NonTechlst
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

NonTechlst2 = []
tokenWord = []
for i in NonTechlst:
    tokenWord.append(word_tokenize(i))
   
   
for i in tokenWord:
     for j in i:
        if j not in stop_words:
            NonTechlst2.append(ps.stem(j.lower()))
        
NonTechlst2
```

```python
X = np.append(np.array(exTechlst),np.array(NonTechlst2))
```

```python
y1 = [1]*len(exTechlst)
y2 = [0]*len(NonTechlst2)
Y = np.append(np.array(y1),np.array(y2))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1425)
X1 = cv.fit_transform(X).toarray()
X1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, Y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
finalClassification = cv.transform(rawSkillslst[1:]).toarray()
y_pred_final = classifier.predict(finalClassification)
print(y_pred_final)
```
